Год релиза можно узнать из открытых источников, заполнив вручную или спарсив данные с сайта. Но это достаточно трудоемкий процесс. Попробуем заполнить пропуски датами релиза с других платформ. Просмотрим 1ю игру.

```python
display(data.query('critic_score.isna()').head(7))
```

На википедии написано, что релиз *Super Mario World* на платформе *Super NES* произошел в 1990 году. А в 2001 вышла версия для платформы *Game Boy Advance*. Остальные игры из этого списка найти на сайте https://www.metacritic.com/ не удалось. Не совпадали либо названия, либо год релиза. Добыть недостающие оценки из открытых источников будет достаточно проблематично. Поэтому просто пропустим их.

А так же из-за ошибок в самих данных (пример с *Super Mario World*, которой просто не выходило в 2001 года на *SNES*).

Отдельно проверим значения tbd в столбце `user_score`.

```python
data.query('user_score=="tbd"').head()
```

На сайте *metacritic* есть пользовательская оценка игры [`Frogger's Adventures: Temple of the Frog`](https://www.metacritic.com/game/game-boy-advance/froggers-adventures-temple-of-the-frog), основанная на 5 отзывах. Но возможно они были написаны уже после создания выборки. При этом оценка критиков полностью совпадает с оценкой на сайте.

Аббревиатура tbd означает to be determined (будет определено). Значит данных по рейтингу нет. Вполне возможно, что никто из пользователей не оценил эти игры.

Заменим tbd на *NaN*, что бы преобразовать столбец в числовой формат.

```python
print('Количество tbd: ',data.query('user_score=="tbd"')['name'].count())
print()
data['user_score'] = data['user_score'].where(data['user_score'] != 'tbd',np.nan)
```

Изменим формат столбцов `year_of_release` на int и user_score на float.

```python
data['year_of_release'] = data['year_of_release'].astype('Int64',errors='ignore')
data['user_score'] = data['user_score'].astype('Float64',errors='ignore')
data.info()
```

## 3 Исследовательский анализ данных
### 3.1 Определение актуального периода

Для начала изучим суммарные продажи всех игр на всех платформах. Построим график.

```python
data.pivot_table(index='year_of_release',aggfunc='count')['name'].plot(title='Динамика количества выпущенных игр по годам',grid=True)
plt.ylabel('Количество выпущенных игры')
plt.xlabel('Год релиза');
```

Пик продаж игр пришелся на 2008-2009 года. Тренд роста начался с 1995 года. По какой-то причине после 2010 года пошел резкий спад.

Для прогнозирования продаж на 2017 год точно не подойдут данные о продажах игр в 20 веке. Нужно исследовать последние тренды. Найдем самые популярные игровые платформы после 2010 года. Запишем их в список.

```python
platforms = data.query('year_of_release>2010').pivot_table(index ='platform',values='all_sales',aggfunc='sum').sort_values(by='all_sales',ascending=False).head(8)
display(platforms)
platforms = platforms.index.tolist()
```

Самые популярные платформы после 2010 года - консоли от SONY и Microsoft предыдущего поколения (PS3 и X360). Следом идут консоли текущего поколения от тех же производителей (PS4 и XOne). Так же в выборке присутствуют *Nintendo 3DS*, PC и Wii с WiiU. Это вполне отражает состояние рынка игровых платформ с 2010 по 2016 год.

Построим графики продаж каждой из топовых платформ по годам.

```python
def popularity_platform(df, platforms): #Напишем функцию, т.к. придется воспользоваться этим кодом в дальнейшем.
    _, ax = plt.subplots(figsize=(8, 8));
    for platrform in platforms:
        (df.query('year_of_release>=2005 and platform==@platrform').pivot_table(index='year_of_release',values='all_sales',aggfunc='sum').plot(ax=ax,))
#Год релиза ограничен 2005м, что бы платформа PC не растягивала график
    plt.ylabel('Продажи, млн. копий')
    plt.xlabel('Год релиза')
    plt.legend(platforms)
    plt.title("Количество проданных копий в год по платформам");
popularity_platform(data, platforms)
```

Исходя из графика, можно сказать, что жизненый цикл игровой платформы длится около 8 лет(с 2006 по 2012 Wii, с 2006 по 2014 для PS3 и X360, c 2011 по 2016 для 3DS). Самые новые платформы - 3DS, XOne и PS4. При этом лидером по продажам в последнее время является PS4. На 2м месте XOne. Также заметно, что новая консоль WiiU не пользуется успехом.

Выберем актуальный период. Пусть это будет год релиза самой популярной в 2016 году консоли (PS4) - 2013 год. То есть будем исследовать данные с 2013 года включительно. В этом же году тренд уменьшения количества продаж видеоигр прекратился.

```python
data_actual = data.query('year_of_release>=2013')
```

### 3.2 Анализ актуального периода
Найдем самые популярные платформы после 2013 года. Также построим график с количеством продаж. На этот раз возьмем только 5 самых популярных платформ.

```python
platforms = (data_actual.
        pivot_table(index ='platform',values='all_sales',aggfunc='sum').sort_values(by='all_sales',ascending=False).head())
display(platforms)
platforms = platforms.index.tolist()
popularity_platform(data_actual, platforms)
```

Ощутимо лидирует по продажам только платформа PS4. Остальные находятся примерно на 1м уровне.

До 2015 года наблюдался рост только у консолей XOne и PS4. Остальные игровые платформы только теряют в проданных копиях игр. Но т.к. 2016 год еще не закончился, возможно ситуация у XOne и PS4 немного улучшится (например, компании раскроют продажи в декабре).

Самая потеницально прибыльная платформа - PS4. На втором месте XOne.

Построим "ящики с усами" по продажам игр на разных топовых платформах.

```python
data_actual[['platform','all_sales']].query('platform in @platforms').boxplot(by='platform', figsize=(6,6), showfliers=False)
plt.title('Распределение числа проданных копий 1й игры на разных платформах')
plt.ylabel('Количество проданных копий, млн');
plt.xlabel('Платформа');
```

Количество проданных копий каждой игры на конкретной платформе зависит от успешности самой игры и популярности платформы.

По ящикам видно, что наибольшее количество копий продается на консоле X360, наименьшее на 3DS. Возможно, на X360 чаще выпускают игры ААА класса, а на 3DS преобладает большое количество инди-игр. Остальные платформы лесенкой расположились между 2 лидерами.

Проверим, как отзывы пользователей и критиков влияют на продажи. Для этого построим диаграммы и посчитаем кореляцию.

```python
for platform in platforms:
    display(f'Платформа {platform}')
    display(data_actual.query('platform == @platform')[['all_sales','critic_score','user_score']].corr()['all_sales'])
    (data_actual.query('platform == @platform')[['all_sales','critic_score','user_score']]
                                .plot(x='critic_score',y='all_sales',kind='scatter',title = f'Платформа {platform}'))
    plt.xlabel('Оценка критиков')
    plt.ylabel('Продажи, млн копий')
    (data_actual.query('platform == @platform')[['all_sales','critic_score','user_score']]
                                .plot(x='user_score',y='all_sales',kind='scatter',title = f'Платформа {platform}'))
    plt.ylabel('Продажи, млн копий')
    plt.xlabel('Оценка пользователей')
    plt.ylim(0,9) #Что бы убрать выбросы
```

Рассмотрим полученные результаты по каждой платформе.
1. PS4. Наблюдается слабая корреляция между продажами и отзывами критиков. При этом продажи и оценка пользователей никак не зависят друг от друга.
2. PS3. Наблюдается слабая корреляция между продажами и отзывами критиков. При этом продажи и оценка пользователей никак не зависят друг от друга.
3. X360. Наблюдается слабая корреляция между продажами и отзывами критиков. При этом продажи и оценка пользователей никак не зависят друг от друга.

### 4.2 Вывод
Северная Америка и Европа представляют собой одинаковый рынок видеоигр. Предпочтиния пользователей из этих регионов во многом схожи. Любимая консоль - PS4, жанры игр - экшн и шутер, возрастное ограничение - для взрослых.

Отдельно стоит Япония. Предпочтения пользователей определяются местными производителями электроники. Чаще всего здесь покупаются ролевые игры для подростков. 

```python
alpha =.05 # зададим уровень значимости
PC = data_actual.query('platform=="PC"')['user_score'].dropna() #Оценки игр на PC
XOne = data_actual.query('platform=="XOne"')['user_score'].dropna() #Оценки игр на XOne

print(f'Среднее значение оценок игр на PC:   {PC.mean():.2f}, стандартное отклонения {PC.std(ddof=1):.2f}, дисперсия {PC.var(ddof=1):.2f}')
print(f'Среднее значение оценок игр на XOne: {XOne.mean():.2f}, стандартное отклонения {XOne.std(ddof=1):.2f}, дисперсия {XOne.var(ddof=1):.2f}')

print('p-значение:',st.levene(PC, XOne).pvalue)
if (st.levene(PC, XOne).pvalue < alpha):
    print("Отвергаем нулевую гипотезу. Дисперсии выборок различаются.")
else:
    print("Не получилось отвергнуть нулевую гипотезу. Дисперсии выборок однородны")
```

Так как дисперсии неоднородны, проведем Т-тест с параметром `equal_var=False`.

```python
results = st.ttest_ind(PC, XOne, equal_var=False) 
print('p-значение:',results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу. Средние рейтинги двух платформ различаются.")
else:
    print("Не получилось отвергнуть нулевую гипотезу. Средние рейтинги 2х платформ одинаковые.")
```

Гипотеза о том, что средние рейтинги платформ PC и XOne одинаковые подтвердилась. Уровень значимости находится на достаточно высоком уровне. Также средние значения и стандартные отклонения примерно равы.
### 5.2 Гипотеза №2
Необходимо проверить следующую гипотезу: "средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные". Все действия будут аналогичны действиям при проверке 1й гипотезы. Сформулируем для этого случая нулевую и альтернативную гипотезы:
* H₀ - средние пользовательские рейтинги жанров Action и Sports равны.
* H₁ - средние пользовательские рейтинги жанров Action и Sports различаются.

Уровень значимости оставим 5%. Найдем p-value.

```python
action = data_actual.query('genre=="Action"')['user_score'].dropna()
sports = data_actual.query('genre=="Sports"')['user_score'].dropna()

print(f'Среднее значение оценок игр жанра Action: {action.mean():.2f}, стандартное отклонения {action.std(ddof=1):.2f}, дисперсия {action.var(ddof=1):.2f}')
print(f'Среднее значение оценок игр жанра Sports: {sports.mean():.2f}, стандартное отклонения {sports.std(ddof=1):.2f}, дисперсия {sports.var(ddof=1):.2f}')

print('p-значение:',st.levene(action, sports).pvalue)
if (st.levene(action, sports).pvalue < alpha):
    print("Отвергаем нулевую гипотезу. Дисперсии выборок различаются.")
else:
    print("Не получилось отвергнуть нулевую гипотезу. Дисперсии выборок однородны")
```

Так как дисперсии неоднородны, проведем Т-тест с параметром `equal_var=False`.

```python
results = st.ttest_ind(action, sports, equal_var=False) 

print('p-значение:',results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу. Средние рейтинги жанров action и sports разные.")
else:
    print("Не получилось отвергнуть нулевую гипотезу. Средние рейтинги жанров action и sports одинаковые.")
```

Гипотеза том, что жанры sports и action имеют разные оценки пользователей подтвердилась. Это видно и по среднему значению и стандартному отклонению выборок.

### 5.3 Вывод
Гипотеза о том, что "средние пользовательские рейтинги платформ Xbox One и PC одинаковые" подтвердилась. Возможно это связано с тем, что производитель Xbox One и самой популярной операционной системой на PC - Windows одинаковый.
